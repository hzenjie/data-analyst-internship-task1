```python
import pandas as pd

# 1. Memuat Dataset Mentah
print("--- 1. Memuat Data ---")
df = pd.read_csv('Mall_Customers.csv')
print(f"Dimensi awal: {df.shape[0]} baris, {df.shape[1]} kolom\n")

# 2. Merapikan Nama Kolom (Uniform Header)
# Mengubah jadi lowercase, menghapus spasi, dan menghilangkan karakter spesial
print("--- 2. Standardisasi Nama Kolom ---")
df.columns = df.columns.str.lower() \
                       .str.replace(' ', '_') \
                       .str.replace('(', '') \
                       .str.replace(')', '') \
                       .str.replace('$', '') \
                       .str.replace('-', '_')
print("Nama kolom baru:", list(df.columns), "\n")

# 3. Cek Nilai Kosong (Missing Values)
print("--- 3. Pengecekan Missing Values ---")
print(df.isnull().sum())
print("Hasil: Tidak ditemukan missing values.\n")

# 4. Cek dan Hapus Data Duplikat
print("--- 4. Pengecekan Data Duplikat ---")
jumlah_duplikat = df.duplicated().sum()
print(f"Jumlah baris duplikat: {jumlah_duplikat}")
if jumlah_duplikat > 0:
    df = df.drop_duplicates()
    print("Duplikat dihapus.")
else:
    print("Hasil: Aman, tidak ada data duplikat.\n")

# 5. Standardisasi Nilai Teks (Gender)
print("--- 5. Standardisasi Nilai Teks ---")
# Memastikan teks gender bersih dari spasi tak terlihat dan seragam
df['gender'] = df['gender'].str.strip().str.title()
print("Kategori gender setelah dibersihkan:", df['gender'].unique(), "\n")

# 6. Memastikan Tipe Data Sudah Sesuai
print("--- 6. Pengecekan Tipe Data ---")
print(df.dtypes)

# 7. Ekspor Dataset yang Sudah Bersih
output_file = 'cleaned_dataset.csv'
df.to_csv(output_file, index=False)
print(f"\n[SUKSES] Dataset bersih berhasil disimpan dengan nama: '{output_file}'")
```
