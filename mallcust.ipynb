```python
import pandas as pd

# 1. Load the Raw Dataset
print("--- 1. Loading Data ---")
df = pd.read_csv('Mall_Customers.csv')
print(f"Initial dimensions: {df.shape[0]} rows, {df.shape[1]} columns\n")

# 2. Standardize Column Names (Uniform Headers)
print("--- 2. Column Name Standardization ---")
df.columns = df.columns.str.lower() \
                       .str.replace(' ', '_') \
                       .str.replace('(', '') \
                       .str.replace(')', '') \
                       .str.replace('$', '') \
                       .str.replace('-', '_')
print("New column names:", list(df.columns), "\n")

# 3. Check for Missing Values
print("--- 3. Checking for Missing Values ---")
print(df.isnull().sum())
print("Result: No missing values found.\n")

# 4. Check and Remove Duplicate Records
print("--- 4. Checking for Duplicate Records ---")
duplicate_count = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count}")
if duplicate_count > 0:
    df = df.drop_duplicates()
    print("Duplicates successfully removed.")
else:
    print("Result: Clean, no duplicate records found.\n")

# 5. Standardize Text Values (Gender Column)
print("--- 5. Text Value Standardization ---")
df['gender'] = df['gender'].str.strip().str.title()
print("Gender categories after cleaning:", df['gender'].unique(), "\n")

# 6. Verify Data Types
print("--- 6. Verifying Data Types ---")
print(df.dtypes)

# 7. Export the Cleaned Dataset
output_file = 'cleaned_dataset.csv'
df.to_csv(output_file, index=False)
print(f"\n[SUCCESS] Cleaned dataset has been saved as: '{output_file}'")
```
